## Create a Classification Report with Evidently

```python
from evidently import Dataset
```

```python
from evidently import DataDefinition, BinaryClassification

data_definition=DataDefinition(
        classification=[BinaryClassification(target="stroke", prediction_labels="prediction")]
    )
```

```python
current_dataset = Dataset.from_pandas(
    curr_data,
    data_definition=data_definition,
)
```

```python
refference_dataset = Dataset.from_pandas(
    reff_data,
    data_definition=data_definition,
)
```

```python
# Define report
classification_report = Report(metrics=[
    ClassificationPreset()
])

# Run comparison
my_report = classification_report.run(
    reference_data=refference_dataset,
    current_data=current_dataset,
    
)
```

```python
my_report.save_html("../evidently_report/model_quality_report.html")
```

```python
import numpy as np
import re

def clean_metrics(obj):
    if isinstance(obj, dict):
        cleaned = {}
        for k, v in obj.items():
            # Special handling for metric_id cleanup
            if k == "metric_id" and isinstance(v, str):
                # Remove parentheses and their contents: "F1Score(conf_matrix=True)" → "F1Score"
                cleaned[k] = re.sub(r"\(.*?\)", "", v).strip()
            else:
                cleaned[k] = clean_metrics(v)
        return cleaned
    elif isinstance(obj, list):
        return [clean_metrics(item) for item in obj]
    elif isinstance(obj, np.generic):
        return obj.item()
    else:
        return obj
```

```python
import mlflow
```

```python
mlflow.set_experiment("classification-monitoring")
with mlflow.start_run(run_name="classification-drift-check"):
    mlflow.log_artifact("../evidently_report/model_quality_report.html", artifact_path="reports")
    # Example: Log a metric from the report
    report_dict = my_report.dict()
    report_dict = clean_metrics(report_dict)
    for metric in report_dict.get("metrics", []):
        name = metric.get("metric_id")
        value = metric.get("value")

        if isinstance(value, dict):
            for k, v in value.items():
                mlflow.log_metric(f"{name}_label_{k}", v)
        else:
            mlflow.log_metric(name, value)
```
